1.Plotting the pie chart

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
slices = [30, 40, 20, 50,10]
```

```python
plt.pie(slices)
plt.show()
```

2.Adding labels to pie chart

```python
labels= ['thirty', 'fourty', 'twenty', 'fifty', 'sixty']
plt.pie(slices,labels=labels)
plt.show()
```

3.Setting edge color

```python
plt.pie(slices, labels=labels, wedgeprops = {'edgecolor':'k', 'linewidth':10})
plt.show()
```

4.Setting color of slices

```python
color = ['red', 'pink','blue','green','yellow']
plt.pie(slices,labels=labels, colors=color, wedgeprops = {'edgecolor':'k', 'linewidth':5})
plt.show()

```

5.Plotting real world data

```python
labels= ['Javascript', 'HTML/CSS', 'SQL', 'Python', 'Java']

slices = [59129, 43424, 31213,24232,11575]

plt.pie(slices,labels=labels, colors=color, wedgeprops = {'edgecolor':'k', 'linewidth':5})
plt.show()
```

6. Exploding the slice

```python
explode = [0,0,0,0.1,0]

plt.pie(slices,labels=labels, explode=explode, shadow=True, colors=color, wedgeprops = {'edgecolor':'cyan', 'linewidth':5})
plt.show()
```

8.Setting the starting scale

```python

plt.pie(slices,labels=labels,explode= explode,shadow=True, startangle= 60,colors=color, wedgeprops = {'edgecolor':'cyan', 'linewidth':5})
plt.show()
```

9.Display percentage of each slides

```python

plt.pie(slices,labels=labels,explode= explode,shadow=True, startangle= 60,autopct= "%0.1f%%", colors=color, wedgeprops = {'edgecolor':'cyan', 'linewidth':5})
plt.show()
```
