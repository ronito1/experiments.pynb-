## *6.Write a prog to create a dataframe using a list of dicts for the sales dataset with atlst 3 columns. Print the dataframe. and delete the 2nd column using pop function*

```python
x = {'one': pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd']),'two': pd.Series([10, 20, 30, 40], index=['a', 'b', 'c', 'd'])}
df = pd.DataFrame(x)
print(df)
#addin a new col to an existing dataframe object with column label by

print("Adding a new column by passing as series")

df['three']=pd.Series([10,20,30],index=['a','b','c'])
print(df)

print('adding a new column using the existing  columns in dataframe:')
df['four']=df['one']+df['three']
print(df)
```

```python


df=pd.DataFrame(df)
print("our dataframe is:")
print(df)
#using del function
print('deleting the first col using DEL function:')
del(df['one'])
print(df)

print("Deleting column using pop")
df.pop('three')
print(df)
```
