1.Creating plots

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
```

```python
#generating 10 ramndom no's btw 25 to 35
numbers = [random.randrange(25,35,1) for numbers in range(10)]
numbers = sorted(numbers, reverse=False)
print(numbers)
```

```python

#generating 10 random no's btw 30k to 45k
numbers1 = [random.randrange(35000,45000,1) for numbers1 in range(10)]
numbers1 = sorted(numbers1, reverse=2)
print(numbers1)
```



# *1.1 Plotting Line Plot*

```python
plt.plot(numbers, numbers1)
```

```python
plt.plot(numbers, numbers1)
plt.title("Median salary in $ with Age")
plt.xlabel("Age")
plt.ylabel("Median salary in $")
plt.legend()
```

# **1.2**

```python
numbers = [random.randrange(25,45,1) for numbers in range(10)]
numbers = sorted(numbers, reverse=False)
print(numbers)
numbers1 = [random.randrange(50000,75000,1) for numbers1 in range(10)]
numbers1 = sorted(numbers1, reverse=2)
print(numbers1)
py_numbers1 = [random.randrange(50000,75000) for py_numbers in range(10)]
py_numbers1 = sorted(py_numbers1, reverse=False)
print(py_numbers1)
```

```python
plt.plot(numbers, numbers1, label = "All developers")
plt.plot(numbers1, py_numbers1,label = "Python developers")
plt.title("Median salary in $ with Age")
plt.xlabel("Age")
plt.ylabel("Median salary in $")
plt.legend()
```

# *1.4 MARKERS*

```python
plt.plot(numbers, numbers1, color = 'k', linestyle = '--', marker = "X", label = "All developers")
plt.plot(numbers1, py_numbers1, color = 'g', linestyle = '-', marker = "o", label = "Python developers" )
plt.title("Median salary in $ with Age")
plt.xlabel("Age")
plt.ylabel("Median salary in $")
plt.legend()
```

# *1.6 Add padding to the plots*

```python
plt.plot(numbers, numbers1, color = "m", linestyle = '--', marker = "X",linewidth='3', label = "All developers")
plt.plot(numbers, py_numbers1, color = "g", linestyle = '-', marker = "o", linewidth='3', label = "Python developers" )
plt.plot(numbers, js_numbers, color = "#FF3355", linestyle = ':', marker = "D", label = "Javascript developers" )

plt.title("Median salary in $ with Age")
plt.xlabel("Age")
plt.ylabel("Median salary in $")
plt.legend()
plt.tight_layout()
```
