```python
def estimate_coeff(p,q):
  n1=np.size(p)
  m_p=np.mean(p)
  m_q=np.mean(q)

  SS_pq=np.sum(q*p)-n1*m_q*m_p
  SS_pp=np.sum(p*p)-n1*m_q*m_p

  b_1=SS_pq/SS_pp
  b_0=m_q-b_1*m_p
  return(b_0,b_1)

def plot_regression_line(p,q,b):
  plt.scatter(p,q,color='m',marker="p",s=30)

  q_pred=b[0]+b[1]*p

  plt.plot(p,q_pred,color='g')

  plt.xlabel('p')
  plt.ylabel('q')
  plt.show()

def main():
  p=np.array(np.random.randint(10,20,size=(10)))
  q=np.array(np.random.randint(500,800,size=(10)))
  b=estimate_coeff(p,q)
  print("estimated coefficients are :\nb_0={}\nb_1={}:".format(b[0],b[1]))

  plot_regression_line(p,q,b)

if __name__ == '__main__':
  main()
```
