Matplot lib

```python
#EX:1 Draw a line in a diagram from position(0,0) to position(6,250):
import matplotlib.pyplot as plt
import numpy as np
```

```python
xpoints = np.array([0,6])
ypoints = np.array([6,250])

plt.plot(xpoints, ypoints)
plt.show()
```

# **Matplot plotting**

# **Plotting without line**

```python
#Draw 2 points in the diagram, one at (1,3) and other at (8,10)


xpoints = np.array([1,8])
ypoints = np.array([3,10])

plt.plot(xpoints, ypoints, '*')
plt.show()
```

# **Multiple points**

```python
#Draw a line in  a diagram from position (1,3) to (2,8) to (6,1) and finally to (8,10):

xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints)
plt.show()
```

# **Matplot lib MARKERS**

```python
ypoints = np.array([8,1,5,7])

plt.plot(ypoints, marker = 'x', color = 'magenta')
plt.show()
```

# Format strings **fmt**

```python
ypoints = np.array([8,1,5,7])

plt.plot(ypoints, 'x:b')
plt.show()
```

draw a line for 4 different coordinate values use the concept of marker, linestyle, and color of lines

```python
xpoints = np.array([2,4])
ypoints = np.array([4,5])
zpoints = np.array([6,2])
lpoints = np.array([6,4])
lpoints = np.array([10,4])

plt.plot(xpoints, 'o:c', linestyle='-')
plt.plot(ypoints, 'h:k',linestyle='--')
plt.plot(zpoints, '*:r',linestyle='-')
plt.plot(lpoints, '^:y',linestyle='--')
plt.plot(lpoints, '*:b',linestyle='-')
plt.show()
```

# **Marker size**

```python
ypoints = np.array([8,2,1,5,7])

plt.plot(ypoints, 'x:k', ms=20)
plt.show()
```

Marker edge color

```python
ypoints = np.array([8,2,1,5,7])

plt.plot(ypoints, 'x:k', ms=20,mec= 'm')
plt.show()
```

```python
ypoints = np.array([8,2,1,5,7])

plt.plot(ypoints, marker = 'o', mec= 'r', mfc= "Black", ms=30)
plt.show()
```
