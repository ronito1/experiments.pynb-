***1.plotting a bar chart***

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [12,24,35,44,55,67,21]
devs_y = [10,20,30,40,50,60,70]
```

```python
plt.bar(x, devs_y, label= "All developers")
plt.xlabel("Ages")
plt.ylabel("Median salary in USD")
plt.title("Median salary by Age")
plt.legend()


```

***5.Plotting the bar plot from pandas dataframe***

```python
data = pd.read_csv("/content/data.csv")
data
```

```python
from collections import Counter
```

```python
ids = data['Responder_id']
language_responses = data['LanguagesWorkedWith']
```

```python
language_counter = Counter()
```

```python
for response in language_responses:
  language_counter.update(response.split(";"))
```

```python
languages = []
popularity = []

for item in language_counter.most_common(15):
  languages.append(item[0])
  popularity.append(item[1])
```

```python
plt.bar(languages, popularity)
plt.xlabel("Languages")
plt.xlabel("Popularity")
plt.title("Language Popularity among developers")
plt.show()
```

## **Plot horizontal Bar chart**

```python
plt.barh(languages, popularity)
plt.xlabel("Popularity")
plt.title("Language Popularity among developers")
plt.show()
```
